````python
qtd_categoria = []

for coluna in data.columns:
    num_categoria = data[coluna].nunique()
    qtd_categoria.append(num_categoria)
    
print(qtd_categoria)
````

````python
metadados['qtd_categoria'] = qtd_categoria
metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
data_dummy = pd.get_dummies(data, columns=['sexo', 'posse_de_imovel', 'tipo_renda', 
    'educacao', 'estado_civil', 'tipo_residencia'],drop_first=True)

data_dummy
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
crosstab_email = pd.crosstab(data['possui_email'], data['mau'])
crosstab_veiculo = pd.crosstab(data['posse_de_veiculo'], data['mau'])
````

````python
crosstab_email[True]
````

````python
crosstab_email['proporcao_mau'] = crosstab_email[True] / (crosstab_email[False] + crosstab_email[True])
crosstab_veiculo['proporcao_mau'] = crosstab_veiculo[True] / (crosstab_veiculo[False] + crosstab_veiculo[True])

print(crosstab_email)
print('\n')
print(crosstab_veiculo)
````

````python
#Resposta:
#De acordo com os dados acima, podemos entender que a variável possui_email tem uma diferença um pouco maior na proporção de maus pagadores entre os grupos comparados à variável posse_de_veiculo.
#Sendo assim, a variável mais poderosa (mesmo que seja por pouco) para prever maus pagadores é possui_email por possuir maior variação
````
